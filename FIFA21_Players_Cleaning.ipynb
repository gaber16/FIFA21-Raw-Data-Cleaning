### Drop the Loan Data End column as it has too many NaN values 17966 from 18979 instances are NaNs

### Converting the height column to ints and changing the feet to cm.
#### When we check for the unique values in the Height column we find that some cells are in the cm format and others are in the feet format. Plus being of object type we have to change either the cm to feet or the feet to cm and change the type of the data to numeric. In this case we will change all to the cm format as it is the standered unit for meacuring Height.

```python
fifa21['Height'].unique()
```

```python
mask = fifa21['Height'].str.endswith('"')
df_feet = fifa21.loc[mask, 'Height']
fifa21['Height'] = fifa21['Height'].str.strip('cm')
fifa21.loc[mask, 'Height'] = df_feet.str.extract(r"(\d*)\'(\d*)").apply(pd.to_numeric, errors='coerce').mul([30.48, 2.54]).sum(axis=1).round().astype(int)
```

```python
fifa21['Height'] = fifa21['Height'].astype(int)
```

```python
fifa21['Height'].unique()
```

### Converting the weight column to floats and changing the pounds to kg.
#### when we check for the unique values in the Weight column we find that some weights are in the kg format and other are in the lbs format. Plus they are all of object type as we can see in the cell bellow. It is also the same case as of the Height's column case. Here we will change all weights to kgs as it is the most common unit for meausring weights.

```python
fifa21['Weight'].unique()
```

```python
mask = fifa21['Weight'].str.endswith('lbs')
fifa21['Weight'] = fifa21['Weight'].str.strip('kglbs')
df_lbs = fifa21[mask].astype(float)
fifa21.loc[mask, 'Weight'] = df_lbs['Weight'] * (1/2.2046)
```

```python
fifa21['Weight'] = fifa21['Weight'].astype(int)
```

```python
fifa21['Weight'].unique()
```

### Now for the value, wage, and release clause columns they are of type object which needs to be changed to numeric.
#### As we see from the below cells that all 3 of those columns contain the M -> million, K -> thousand, and Euro symbol. We will strip the Euro symbol, change to numeric and multiply by million or thousand to get the desired value for each instance.

```python
fifa21['Value'].unique()
```

```python
fifa21['Wage'].unique()
```

```python
fifa21['Release Clause'].unique()
```

```python
def money_columns_cleaning(df):
    df = df.str.strip('€')
    symbol_to_value = {'M':'*1000000', 'K': '*1000'}
    conversion = df.replace(symbol_to_value, regex= True).map(pd.eval)
    df = conversion.astype(int)
    return df

fifa21['Value'] = money_columns_cleaning(fifa21['Value'])
fifa21['Wage'] = money_columns_cleaning(fifa21['Wage'])
fifa21['Release Clause'] = money_columns_cleaning(fifa21['Release Clause'])
```

```python
print(fifa21['Value'].unique()[:5])
print(fifa21['Wage'].unique()[:5])
print(fifa21['Release Clause'].unique()[:5])
```

### Conversion of skill stars to number as ordinal data

```python
cols = ['W/F', 'SM', 'IR']
print(fifa21['W/F'].unique())
print(fifa21['SM'].unique())
print(fifa21['IR'].unique())
```

```python
fifa21[cols] = fifa21[cols].apply(lambda x: x.str.replace('★', ''))
fifa21[cols] = fifa21[cols].apply(lambda x: x.str.replace(' ', ''))
print(fifa21['W/F'].unique())
print(fifa21['SM'].unique())
print(fifa21['IR'].unique())
fifa21[cols] = fifa21[cols].astype(int)    
fifa21[cols].dtypes
```

### The Hits column had NaN values and numeric values but with K -> thousand.
#### So we removed the K and multiplied by 1000 and changed the column type to int

```python
# mask = fifa21['Hits'].str.endswithwith('K')
fifa21['Hits'].fillna(0, inplace= True)
symbol_to_value = {'K': '*1000'}
fifa21['Hits'] = fifa21['Hits'].replace(symbol_to_value, regex= True).map(pd.eval).astype(int)
```

```python
fifa21['Hits'].unique()
```
